```python
N = 8

def matrix(N):
    count = 1
    matrix_A = []
    matrix_B = []
    for i in range(N):
        temp = []
        for j in range(N):
            temp.append(count)
            count += 1
        matrix_A.append(temp)
        matrix_B.append(temp)

    result_matrix = []
    for i in range(N):
        temp_list = []
        for j in range(N):
            temp = 0
            for k in range(N):
                temp += matrix_A[i][k] * matrix_B[k][j]
            temp_list.append(temp)
        result_matrix.append(temp_list)
    
    for i in range(N):
        for j in range(N):
            print(result_matrix[i][j], end=" ")
        print()
```

```python
matrix(4)
```

```python
matrix(8)
```
